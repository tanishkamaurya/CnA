### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
median_mpg = df['mpg'].median()
above_median_mpg = df[df['mpg'] > median_mpg].copy()
above_median_mpg['hp_per_weight'] = above_median_mpg['horsepower'] / above_median_mpg['weight']
max_ratio = above_median_mpg['hp_per_weight'].max()
best_cars = above_median_mpg[above_median_mpg['hp_per_weight']==max_ratio]
print("Cars with best horsepower-to-weight ration and above median mpg:")
print(best_cars['name'].values)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
import matplotlib.pyplot as plt
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().unstack()
avg_mpg.plot(marker ='o')
plt.title('Average MPG over years by origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.grid(True)
plt.legend(title = 'Origin')
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
# Step 1: Count how many years each car model appears in
name_counts = df['name'].value_counts()
repeated_names = name_counts[name_counts > 1].index

# Step 2: Filter df to keep only repeated car names
repeated_df = df[df['name'].isin(repeated_names)]

# Step 3: Group by 'name' and compute stats
grouped = repeated_df.groupby('name').agg(
    count=('model_year', 'nunique'),
    appearances=('name', 'count'),
    mpg_std=('mpg', 'std'),
    mpg_mean=('mpg', 'mean')
).reset_index()

# Step 4: Filter for std < 1.0 and at least 2 model years
consistent = grouped[(grouped['count'] >= 2) & (grouped['mpg_std'] < 1.0)]

# Step 5: Sort by appearances (desc), then avg mpg (desc)
consistent_sorted = consistent.sort_values(by=['appearances', 'mpg_mean'], ascending=[False, False])

# Step 6: Display results
print("Consistent car models (std < 1.0 mpg, multiple years):")
print(consistent_sorted[['name', 'appearances', 'mpg_mean']])
```
